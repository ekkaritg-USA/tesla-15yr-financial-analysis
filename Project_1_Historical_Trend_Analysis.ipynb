* **Charging Infrastructure**: Global fast-charging networks expanded from virtually non-existent to over 1.5 million stations
* **Manufacturing Innovation**: Tesla pioneered mega-casting and structural battery pack technologies, reducing production complexity

### Regulatory Environment

* **Emissions Standards**: Increasingly stringent global regulations accelerated EV adoption
* **Government Incentives**: Various subsidies, tax credits, and rebates significantly influenced EV market dynamics
* **Policy Shifts**: The Inflation Reduction Act (USA) and European Green Deal created major regional market impacts
* **Changing Mandates**: Several markets announced future bans on internal combustion engine vehicles, signaling a definitive industry transition

### Capital Markets Context

* **Valuation Evolution**: Tesla's market capitalization increased from approximately 1.7 billion in 2010 to periods exceeding 1 trillion
* **Investment Influx**: EV and battery technology attracted over 400 billion in global investment between 2020-2024
* **SPAC Phenomenon**: The 2020-2022 period saw numerous EV startups enter public markets through special purpose acquisition companies

````python
# @title
import plotly.express as px
import plotly.graph_objects as go
import pandas as pd
import numpy as np

# Create the enhanced revenue plot
fig_revenue = px.line(tesla_df,
                     x=tesla_df.index,
                     y='Total Revenue',
                     markers=True)

# Enhance with a more polished look
fig_revenue.update_traces(
    line=dict(width=3, color='#E31937'),  # Tesla red color, thicker line
    marker=dict(size=8, color='#E31937', line=dict(width=2, color='white'))
)

# Improve layout with Tesla branding and better formatting
fig_revenue.update_layout(
    title={
        'text': 'Tesla Total Revenue Growth (2010-2024)',
        'font': {'size': 24, 'family': 'Arial, sans-serif', 'color': '#333333'},
        'x': 0.0,
        'y': 0.95
    },
    paper_bgcolor='white',
    plot_bgcolor='#f8f8f8',
    xaxis=dict(
        title='Year',
        title_font=dict(size=16),
        tickfont=dict(size=14),
        showgrid=False,
        zeroline=False
    ),
    yaxis=dict(
        title='Revenue (Millions USD)',
        title_font=dict(size=16),
        tickfont=dict(size=14),
        showgrid=False
    ),
    margin=dict(l=80, r=80, t=100, b=80),
    showlegend=False,
    hovermode='x unified'
)

# Add hover template for better data viewing
fig_revenue.update_traces(
    hovertemplate='<b>Year:</b> %{x}<br><b>Revenue:</b> $%{y:,.0f} million<extra></extra>'
)

# Optional: Add trend line
z = np.polyfit(range(len(tesla_df.index)), tesla_df['Total Revenue'], 1)
p = np.poly1d(z)
fig_revenue.add_trace(
    go.Scatter(
        x=tesla_df.index,
        y=p(range(len(tesla_df.index))),
        mode='lines',
        line=dict(color='rgba(0,0,0,0.2)', width=2, dash='dash'),
        name='Trend',
        showlegend=False
    )
)

# Show the plot
fig_revenue.show()
````

* **Growth Phase (2013-2017):** Revenue expanded at a steady but significant rate from 2 billion to nearly 12 billion, representing a CAGR of 42.4% during the premium vehicle era.
* **Hyper-Growth Phase (2018-2022):** Revenue accelerated dramatically from 21.5 billion to 81.5 billion, achieving a remarkable CAGR of 39.5% as Model 3/Y production scaled globally.
* **Maturation Phase (2023-2024):** Growth decelerated significantly with revenue increasing only 0.9% in 2024, suggesting market saturation and intensifying competition.
